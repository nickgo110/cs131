```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("https://raw.github.com/nickgo110/cs131/main/imdb_anime.csv")
```

```python
data.head()
```

```python
data['Number of Votes'] = data['Number of Votes'].astype(str)
data['Genre'] = data['Genre'].astype(str)
```

```python
data.dtypes
```

```python
data.head()
```

```python
data['Number of Votes'] = data['Number of Votes'].str.replace(',', '')
```

```python
data.head()
```

```python
data['Number of Votes'] = pd.to_numeric(data['Number of Votes'], errors='coerce')
data['Gross'] = pd.to_numeric(data['Gross'], errors='coerce')
```

```python
data['Number of Votes'] = data['Number of Votes'].fillna(0).astype(int)
data['Gross'] = data['Gross'].fillna(0).astype(int)
```

```python
data.dtypes
```

```python
data.to_csv('japanese_anime.csv', index=False)
```

```python
print(data['Number of Votes'].sum())
```

```python
# find sum of number of votes throughout all the years
total1 = data.loc[data['Year'] == '(1980)', 'Gross'].sum()
total2 = data.loc[data['Year'] == '(1990)', 'Gross'].sum()
total3 = data.loc[data['Year'] == '(2000)', 'Gross'].sum()
total4 = data.loc[data['Year'] == '(2010)', 'Gross'].sum()

# set x-axis to the years and y-axis to total gross
x_vals = [1980, 1990, 2000, 2010]
y_vals = [total1, total2, total3, total4]

# label the graph
plt.title("Year vs Gross")
plt.xlabel('Years')
plt.ylabel('Gross')

# set a variable which includes the line
line = plt.plot(x_vals, y_vals)

# change the way the line looks
plt.setp(line, color='g', linewidth=3.0)

# plot the graph
plt.show()
```
